Testing of downloading the data from the dataset

```python
#import required libraries

import os
import dotenv
import gdown
```

```python
#load the secret files

dotenv.load_dotenv()
dataset_url = os.getenv('dataset_path')
dataset_output = os.getenv('dataset_download_path')

print(dataset_output)
```

```python
#Check if storage file is there

print(os.path.exists(dataset_output))
```

```python
#Try to load the dataset documents using gdown

try:
    gdown.download_folder(
        url = dataset_url,
        output = dataset_output,
        quiet = False,
        use_cookies = True
    )

    if os.path.exists(dataset_output) and os.listdir(dataset_output):
        print(f'Download successful. {len(os.listdir(dataset_output))} files uploaded.') 
    else:
        raise Exception('Please double check the path to the dataset dump folder')

except Exception as e:
    print(f'An error occured : {e}')
```
