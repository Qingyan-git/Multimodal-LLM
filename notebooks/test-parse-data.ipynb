Testing on using PyMuPdf to parse the downloaded pdfs

```python
#import required libraries

import pymupdf
import pathlib
import os
import dotenv
```

```python
#get some env variables

dotenv.load_dotenv()
test_file_path = os.getenv('test_file_path')
test_file_output = os.getenv('test_file_output')
```

```python
doc = pymupdf.open(test_file_path) # open a document
doc_title = doc.metadata.get('title','No title found')
print(doc_title)

output_path = os.path.join(test_file_output, f'{doc_title}.txt')

with open(output_path, 'w', encoding='utf-8') as f:
    for page in doc:
        text = page.get_text()
        f.write(text)

print('PDF textual data saved as extracted data')
```
